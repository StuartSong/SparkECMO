```python
# Load data
action_space = pd.read_csv("ECMO_action_space.csv")
score_space = pd.read_csv("ECMO_score_space.csv")
data_clustered = pd.read_csv("ECMO_data_clustered.csv")
unique_action_space = pd.read_csv('../unique_action_space.csv')

# Merge data
merge_data = pd.DataFrame()
merge_data["csn"] = data_clustered.csn
merge_data["action"] = action_space.action_number
merge_data["short_term_reward"] = score_space.SOFA_sigma
merge_data["long_term_reward"] = np.where(score_space['mortality_in_hospital'] == 0, 1, -1)
merge_data["mortality_flag"] = score_space.mortality_in_hospital

from sklearn.model_selection import train_test_split
unique_csn = data_clustered['csn'].unique()
train_csn, test_csn = train_test_split(unique_csn, test_size=0.2, random_state=3)

train_cluster = data_clustered[data_clustered['csn'].isin(train_csn)]
test_cluster = data_clustered[data_clustered['csn'].isin(test_csn)]
train_cluster.to_csv("ECMO_data_clustered_train.csv", index= False)
test_cluster.to_csv("ECMO_data_clustered_test.csv", index= False)

train_merge_data = merge_data[merge_data['csn'].isin(train_csn)]
test_merge_data = merge_data[merge_data['csn'].isin(test_csn)]
train_merge_data.to_csv("merge_data_train.csv", index= False)
test_merge_data.to_csv("merge_data_test.csv", index= False)
```
