```python
import pandas as pd
import numpy as np
from tableone import TableOne
```

```python
df = pd.read_csv('ECMO_data_states_actions_scores_full_table_one.csv')
df
```

```python
df['vent_start_time'] = pd.to_datetime(df['vent_start_time'])
df['vent_stop_time'] = pd.to_datetime(df['vent_stop_time'])

# Calculate LOS in hours
df['LOS_hours'] = (df['vent_stop_time'] - df['vent_start_time']).dt.total_seconds() / 3600

# Calculate LOS in days
df['LOS_days'] = df['LOS_hours'] / 24

df['LOS_days'] 
```

```python
# Relevant features for Table 1
demographics = ['age', 'gender', 'race','daily_weight_kg']
other = ['LOS_days']
sofas = ['SOFA_renal','SOFA_cardio','SOFA_resp']
actions = ['peep','fio2','vent_rate_set']

# Define columns to include in Table 1
selected_features = demographics + other +sofas

# Step 1: Aggregate time series data to patient level
# Averaging continuous variables and taking first valid entry for categorical ones
def aggregate_patient_data(df):
    grouping_col = 'csn'
    
    # Ensure df includes 'csn' for grouping, but don't treat it as numeric or categorical
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    numeric_cols = [col for col in numeric_cols if col != grouping_col]
    
    categorical_cols = df.select_dtypes(exclude=['number']).columns.tolist()
    categorical_cols = [col for col in categorical_cols if col != grouping_col]
    
    agg_funcs = {col: 'mean' for col in numeric_cols}  
    for col in categorical_cols:
        agg_funcs[col] = 'first'
    
    # Group by 'csn' and aggregate
    aggregated = df.groupby(grouping_col).agg(agg_funcs).reset_index()
    return aggregated

aggregated_data = aggregate_patient_data(df[selected_features + ['csn']])
aggregated_data.drop(columns="csn",inplace=True)

# Step 2: Generate Table 1
# Categorical columns and groupby setting (if needed)
categorical = ['gender', 'race']
groupby = None  # Set this to a variable like 'outcome' if you need comparison groups

# Create Table 1
table1 = TableOne(aggregated_data, categorical=categorical, groupby=groupby, pval=False, missing=False)

# Display the Table 1 summary
table1.to_csv('table1_summary.csv')
print(table1)
```
